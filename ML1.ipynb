```python
import pandas as pd
frame = pd.read_csv('insurance.csv')
print(frame.describe())
print(frame.info())
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
fig = px.histogram(frame,x='age',marginal='box',nbins=47,title='Distribution of age')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig2 = px.histogram(frame,x='bmi',marginal='box',title='Distribution of bmi')
fig2.update_layout(bargap=0.1)
fig2.show()
```

```python
fig3 = px.histogram(frame,x='charges',marginal='box',color='smoker',title='Distribution of charges')
fig3.update_layout(bargap=0.1)
fig3.show()
```

```python
fig4 = px.histogram(frame,x='smoker',marginal='box',color='sex',title='Distribution of smokers')
fig4.update_layout(bargap=0.1)
fig4.show()
```

```python
fig5 = px.scatter(frame,x='age',y='charges',color='smoker',opacity=0.8,hover_data=['sex'],title='Age vs Charges')
fig5.update_traces(marker_size=5)
fig5.show()
```

```python
frame2 = pd.read_csv('insurance.csv')
smoker_values = {'yes':1,'no':0}
smoker_numeric = frame2.smoker.map(smoker_values)
frame2['smoker'] = smoker_numeric
sex_map = {'male':1,'female':0}
sex_numeric = frame2.sex.map(sex_map)
frame2['sex'] = sex_numeric
region_map = {'northeast':1,'northwest':2,'southeast':3,'southwest':4}
frame2['region'] = frame2.region.map(region_map)
print(frame2)
import seaborn as sns
sns.heatmap(frame2.corr(),annot=True)
plt.show()
```
